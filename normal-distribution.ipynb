```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Define the range for the x-axis
x = np.linspace(-5, 5, 1000)
# Calculate the PDF of the standard normal distribution
pdf = norm.pdf(x, 0, 1)
# Plotting
plt.plot(x, pdf)
plt.title('Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

```python
# Define the range for the x-axis
x = np.linspace(-5, 5, 1000)
# Calculate the PDF of the standard normal distribution
pdf = norm.pdf(x, 0, 1)
# Plotting
plt.plot(x, pdf)
plt.title('Standard Normal Distribution (Log Scale)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.yscale('log')  # Set the y-axis to a logarithmic scale
plt.grid(True)
plt.show()
```

```python
# Define the range for the original x-axis
x = np.linspace(-5, 5, 1000)
# Calculate the PDF of the standard normal distribution
pdf = norm.pdf(x, 0, 1)
# Adjust x-values to be squared
squared_x = np.sign(x) * (x ** 2)
# Plotting
plt.plot(squared_x, pdf)
plt.title('Standard Normal Distribution (Log y and x^2 Scaled)')
plt.xlabel('x^2')
plt.ylabel('Probability Density')
plt.yscale('log')
plt.grid(True)
plt.show()
```
